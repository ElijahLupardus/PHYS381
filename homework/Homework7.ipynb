```python
#Initial to Final Energy Ratio

def calc_energy_ratio(df):
    
    initial_energy = df['E'].iloc[0]
    final_energy = df['E'].iloc[-1]
    
    return initial_energy/final_energy
    
```

```python
def calc_stuff_from_file(filename):
    
    values_dict = dict()
    
    df = pd.read_csv(filename)
    initial_energy = df['E'].iloc[0]
    final_energy = df['E'].iloc[-1]
    df['T/E'] = df['T']/df['E']
    values_dict['energy_ratio'] = initial_energy/final_energy
    
    medians = df.apply(np.median)
    values_dict['p0_median'] = medians['p_0(t)']
    values_dict['p1_median'] = medians['p_1(t)']
    values_dict['theta_0_median'] = medians['theta_0(t)']
    values_dict['thetadot_0_median'] = medians['thetadot_0(t)']
    values_dict['theta_1_median'] = medians['theta_1(t)']
    values_dict['thetadot_1_median'] = medians['thetadot_1(t)']
    values_dict['T/E'] = medians['T/E']
    
    values_dict['file']= filename
    
    return values_dict
```

```python
#glob will allow us to read multiple csv files at once
import glob
```

```python
# asterisk is a wildcard to tell the code to read any file ending in .csv
csvfiles = glob.glob("/home/cmckay/clusters/CQM/*.csv")
```

```python
calc_stuff_from_file(csvfiles[0])
```

```python
energy_ratios = []

for filename in csvfiles:
    energy_ratios.append(calc_stuff_from_file(filename))
```

```python
datavalues = pd.DataFrame(energy_ratios)

datavalues.to_csv('HW7data.csv')
```
